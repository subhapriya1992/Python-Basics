2. What is an interpreter in Python ?
- In Python, an interpreter is the program that reads and executes Python code line by line, rather than compiling the entire code at once like some other languages(e.g, C or Java).

3. What are pre-defined keywords in Python ?
- Predefined keywords in Python are reserved words that have special meaning to the interpreter. You can not use them as variable names , function names or identifiers because they are part of Python's syntax and structure.
Examples of keywords include if, else , while, for, def etc.

4. Can keywords be used as variable names?
- No, keywords cannot be used as variable names in Python.

5. What is mutability in Python ?
- Mutability refers to whether an object's value can be changed after it is created. Mutable objects can be changed in place ( without changing their identity). Mutable objects, such as lists and dictionaries, can be modified after creation.

6. Why are lists mutable, but tuples are immutable ?
- Lists are meant for collections where you expect to change, grow, or shrink the data.
Tuples are designed for fixed collections of items. Once created, they should not change- this makes them safer and more predictable, especially for use as dictionary keys or in sets.

7. What is the difference between “==” and “is” operators in Python ?
- The "==" and "is" operators in Python are both used for comparison, but they check for different things:
"==" checks if the values of two objects are equal.It compares the content or data held by the objects.
"is" checks if two variables refer to the same object in memory. It compares the memory addresses of the objects.

9. What is type casting in Python ?
- Type casting, also referred to as type conversion, is the process of changing one data type to another in Python.
Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.

10. What is the difference between implicit and explicit type casting ?
- Implicit Type Casting
Automatic conversion done by the compiler/interpreter.
Happens when converting smaller data types to larger data types (e.g., int to float).
Explicit Type Casting
Manual conversion done by the programmer using a cast operator.
Used when converting larger to smaller types or incompatible types.

11. What is the purpose of conditional statements in Python ?
- Conditional statements in Python are used to make decisions in a program based on whether a condition is true or false.
The main purpose is to control the flow of execution by allowing the program to take different paths depending on certain conditions.

12. How does the elif statement work ?
- The elif statement (short for "else if") is used to check multiple conditions after an initial if condition.
Python evaluates conditions from top to bottom. As soon as it finds a condition that is True, it executes the corresponding block and skips the rest.If none of the if or elif conditions are true, the else block (if present) runs.

13. What is the difference between for and while loops ?
- Both for and while loops are used for repeating code, but they are used in different situations based on how and when the loop should end.

for loop > Used when you know ahead of time how many times to repeat.
Often used to iterate over sequences like lists, strings, or ranges.

while loop > Used when the number of repetitions is not known in advance.
Runs as long as a condition remains True.

14. Describe a scenario where a while loop is more suitable than a for loop.

-  A while loop is more suitable than a for loop when the number of iterations is not known in advance, such as waiting for correct user input.

correct_password = "python123"
user_input = input("Enter password: ")

while user_input != correct_password:
    print("Incorrect password. Try again.")
    user_input = input("Enter password: ")

print("Access granted!")

Why while is better here:
You don’t know how many attempts the user will take.
The loop should continue until the user enters the correct password.
A for loop would require setting an arbitrary limit (e.g., max 3 attempts), which doesn't fit this case unless explicitly desired.

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

n = int(input("Enter the Number:"))
product = 1
while n > 1:
  product = product * n
  n = n - 1
print (f"Factorial is : {product}")
```
